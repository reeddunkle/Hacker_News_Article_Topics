```python
import pandas as pd
from goose import Goose
import numpy as np
import lda
from wordcloud import WordCloud
from exceptions import TypeError, RuntimeError

g = Goose()
```

```python
df = pd.read_csv("articles.csv", index_col=False)
```

```python
def get_content(html):
    return g.extract(raw_html=html).cleaned_text
```

```python
html_list = df["html"].tolist()
```

```python
title_list = df['title'].tolist()
```

```python
def clean_html(skip=[]):
    html_clean = []

    i = 0
    for html in html_list:
        print i
        try:
            if i not in skip:
                html_clean.append(get_content(html))

        except (IndexError, TypeError):
            pass
        
        except RuntimeError:
            skip.append(i)
            pass

        i += 1
        
    return html_clean
        
html_clean = clean_html()
```

```python
import string
printable = set(string.printable)
exclude = set(['!', '"', '%', '$', "'", '&', ')', '(', '*', '-', ',', '/', '.', ';', ':', '<', '?', '>', '@', '[', ']', '\\', '_', '^', '`', '{', '}', '|', '~'])
clean_text = []

def clean(text):
    new_s = filter(lambda c: c in printable, text)
    clean_s = ""
    for i, c in enumerate(new_s):
        
        if (c in exclude) and (i < len(new_s) - 1):
            if new_s[i+1] != " ":
                clean_s += c
        elif i == len(new_s):
            if c not in exclude:
                clean_s += c
        else:
            clean_s += c
    
    return clean_s
    

for text in html_clean:
    good_text = clean(text)
    clean_text.append(good_text)
```
